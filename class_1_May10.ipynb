```python
import pandas as pd

# add the "r" before the path to change the special character
df = pd.read_csv(r"\Users\samue\OneDrive\Desktop\MBAN6110S_Data science\messy_data.csv")

# examine the first few lines
print(df.head())

# check whehter the data type is correpondant to the column nature
print(df.info())

# describe a summary of statistics
print(df.describe())

# describe a summary of statistics inclduing categorical variable
print(df.describe(include = 'all'))
```

```python
# data cleansing - filling in missing data

df[df['Product'].isnull()]

# statistics on a subset of data
df[df['Product'] == 'D'].describe()
```

```python
df2 = df.copy()

# setting the missing value as mean
df2['Income'] = df2['Income'].fillna(df['Income'].mean())

# show no more missing value!
df2[df2['Income'].isnull()]
```

```python
df['Income'] = df['Income'].fillna(df.groupby('Gender')['Income'].transform('mean'))

df[df['Income'].isnull()]
```

```python
# drop duplicate
df.drop_duplicates()

# make the drop duplicate as permanent change
df.drop_duplicates(inplace=True)
```
